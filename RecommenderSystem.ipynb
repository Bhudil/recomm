```python
import numpy as np
import pandas as pd
```

```python
columns_names=['user_id','item_id','rating','timestamp']
df=pd.read_csv('u.data',sep='\t',names=columns_names)
df.head()
```

```python
titles=pd.read_csv('Movie_Id_Titles')
titles.head()
```

```python
df=pd.merge(df,titles,on='item_id')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['no of ratings']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
ratings['no of ratings'].hist(bins=50)
```

```python
ratings['rating'].hist(bins=50)
```

```python
sns.jointplot(x='rating',y='no of ratings',data=ratings,alpha=0.6)
```

```python
mat=df.pivot_table(index='user_id',columns='title',values='rating')
mat.head()
```

```python
ratings.sort_values('no of ratings',ascending=False).head()
```

```python
SW_rt=mat['Star Wars (1977)']
LL_rt=mat['Liar Liar (1997)']
```

```python
SW_rt.head()
```

```python
sim2SW=mat.corrwith(SW_rt)
```

```python
sim2LL=mat.corrwith(LL_rt)
```

```python
corr_SW=pd.DataFrame(sim2SW,columns=['Correlation'])
corr_SW.dropna(inplace=True)
corr_SW.head()
```

```python
corr_SW.sort_values('Correlation',ascending=False).head()
```

```python
corr_SW=corr_SW.join(ratings['no of ratings'])
corr_SW.head()
```

```python
corr_SW[corr_SW['no of ratings']>100].sort_values('Correlation',ascending=False).head()
```

```python
corr_LL=pd.DataFrame(sim2LL,columns=['Correlation'])
corr_LL.head()
```

```python
corr_LL.dropna(inplace=True)
```

```python
corr_LL=corr_LL.join(ratings['no of ratings'])
```

```python
corr_LL[corr_LL['no of ratings']>100].sort_values('Correlation',ascending=False).head()
```
